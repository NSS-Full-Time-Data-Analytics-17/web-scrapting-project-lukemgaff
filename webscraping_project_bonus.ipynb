```python
import requests
import pandas as pd
import re
```

```python
URL = 'https://www.billboard.com/charts/hot-100/'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64'}

response = requests.get(URL, headers = headers)
```

```python
from bs4 import BeautifulSoup as BS
```

1. Navigate to https://www.billboard.com/charts/hot-100/. Using BeautifulSoup, extract out the This Week, artist, song, Last Week, Peak Position, and Weeks on Chart values into a pandas DataFrame. Hint: The HTML for the number one ranked song is slightly different from that of the rest of the songs.

```python
soup = BS(response.text, 'html.parser')
```

```python
rows = []
```

```python
items = soup.select('li.o-chart-results-list__item')
```

```python
for item in items:
    title_tag = item.select_one('h3#title-of-a-story')
    if not title_tag:
        continue
    title = title_tag.get_text(strip=True)
    artist_tag = item.select_one('span.c-label')
    artist = artist_tag.get_text(strip=True) if artist_tag else None
    stat_tags = item.select('span.c-label.a-font-primary-s')
    stats = [s.get_text(strip=True) for s in stat_tags]
    last_week  = stats[0] if len(stats) > 0 else None
    peak = stats[1] if len(stats) > 1 else None
    weeks = stats[2] if len(stats) > 2 else None
    rows.append({
        'This Week': len(rows) + 1,
        'Song': title,
        'Artist': artist,
        'Last Week': last_week,
        'Peak Position': peak,
        'Weeks on Chart': weeks})
```

```python
top_songs = pd.DataFrame(rows)
```
